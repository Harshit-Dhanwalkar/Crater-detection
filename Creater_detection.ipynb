```python
import cv2
import numpy as np
from skrobot.planner import rrt_star
import matplotlib.pyplot as plt
```

```python
def detect_craters(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    _, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    crater_positions = []
    for contour in contours:
        M = cv2.moments(contour)
        if M["m00"] != 0:
            cx = int(M["m10"] / M["m00"])
            cy = int(M["m01"] / M["m00"])
            crater_positions.append((cx, cy))

    return crater_positions
```

```python
def rrt_star_planning(start, goal, obstacles, max_iter=500):
    planner = rrt_star.RRTStar(start, goal, obstacles, max_iter=max_iter)
    path, _ = planner.plan()
    return path
```

```python

if __name__ == "__main__":
    image_path = 'your_image_path.jpg'
    crater_positions = detect_craters(image_path)
    print("Crater positions:", crater_positions)

    start_position = (10, 10)
    goal_position = (190, 190)

    # Run RRT* for real-time path planning
    path = rrt_star_planning(start_position, goal_position, crater_positions)

    if path:
        print("Optimized Path:", path)
        
        # Visualization (optional)
        image = cv2.imread(image_path)
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.plot(start_position[0], start_position[1], 'go', markersize=10, label='Start')
        plt.plot(goal_position[0], goal_position[1], 'ro', markersize=10, label='Goal')
        plt.plot(*zip(*path), 'b-', linewidth=2, label='Optimized Path')
        plt.legend()
        plt.show()
    else:
        print("No path found.")
```

```python
image_path = "Craters.jpg"
```

```python
def detect_craters(image_path):
    # Read the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply image processing (e.g., thresholding) to highlight craters
    _, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

    # Find contours of the craters
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Extract crater positions from contours
    crater_positions = []
    for contour in contours:
        M = cv2.moments(contour)
        if M["m00"] != 0:
            cx = int(M["m10"] / M["m00"])
            cy = int(M["m01"] / M["m00"])
            crater_positions.append((cx, cy))

    return crater_positions
```

```python
def draw_craters(image, crater_positions):
    # Draw circles around the detected craters
    for position in crater_positions:
        cv2.circle(image, position, 10, (0, 255, 0), 2)  # Green circle with a radius of 10

    return image
```

```python

if __name__ == "__main__":
    # Detect craters in the image
    crater_positions = detect_craters(image_path)
    print("Crater positions:", crater_positions)

    # Read the original image
    original_image = cv2.imread(image_path)

    if original_image is not None:
        # Draw craters on the original image
        image_with_craters = draw_craters(original_image.copy(), crater_positions)

        # Display the original image
        cv2.imshow("Original Image", original_image)

        # Display the image with detected craters
        cv2.imshow("Image with Craters", image_with_craters)

        cv2.waitKey(0)
        cv2.destroyAllWindows()
    else:
        print(f"Error: Unable to read the image from {image_path}. Please check the file path.")
```
